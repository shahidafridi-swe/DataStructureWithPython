```python
class Book:
    def __init__(self, id_number, title, author):
        self.id_number = id_number
        self.title = title
        self.author = author
        self.is_available = True  # Availability status


class Member:
    def __init__(self, id_number, name):
        self.id_number = id_number
        self.name = name
        self.borrowed_books = []  # List of borrowed books


class Library:
    def __init__(self, id_number, name):
        self.id_number = id_number
        self.name = name
        self.books = []  # List to store book objects
        self.borrowed_books = {}  # Maps books to members

    def add_book(self, book_id, book_title, author):
        new_book = Book(book_id, book_title, author)
        self.books.append(new_book)  # Add book to the library's collection
        print(f"Book '{book_title}' by {author} added to the library.")

    def issue_book(self, member, book_title):
        # Find the desired book
        for book in self.books:
            if book.title == book_title:
                desire_book = book
                break
        else:
            print(f"'{book_title}' is not found in the library.")
            return

        # Check availability
        if desire_book.is_available:
            # Check if the member has already borrowed 3 books
            if len(member.borrowed_books) >= 3:
                print(f"{member.name} has already borrowed three books.")
                return

            # Issue the book
            desire_book.is_available = False
            member.borrowed_books.append(desire_book)
            self.borrowed_books[desire_book] = member
            print(f"{member.name} borrows '{desire_book.title}'.")

        else:
            print(f"'{desire_book.title}' is not available at the moment.")

    def return_book(self, member, book_title):
        # Find the desired book
        for book in member.borrowed_books:
            if book.title == book_title:
                desire_book = book
                break
        else:
            print(f"'{book_title}' is not a valid borrowed book.")
            return

        # Return the book
        desire_book.is_available = True
        member.borrowed_books.remove(desire_book)
        del self.borrowed_books[desire_book]
        print(f"'{desire_book.title}' has been returned to the library.")
```
